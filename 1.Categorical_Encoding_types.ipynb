# Categorical Encoding Cheat Sheet

---

## 1. Label Encoding
* **What it does:** Converts each category into an integer from $0$ to $N-1$.
* **When to use:** Only on the Target variable ($y$).
* **Warning:** Do not use on input features ($X$) because models treat numbers like math (e.g., assumes $2 > 1$).

---

## 2. Ordinal Encoding
* **What it does:** Maps categories to numbers using a custom, predefined rank.
* **When to use:** On Ordinal features with a clear natural sequence (e.g., `Low` < `Medium` < `High`).
* **Syntax:** `OrdinalEncoder(categories=[['Low', 'Medium', 'High']])`

---

## 3. One-Hot Encoding (`sklearn`)
* **What it does:** Creates a separate $0/1$ binary column for every category.
* **When to use:** On Nominal features without any order (e.g., `City`, `Color`).
* **Best Practice:** Always use in ML pipelines and set `drop='first'` to prevent the Dummy Variable Trap.

---

## 4. Pandas `get_dummies()`
* **What it does:** Quick built-in function for One-Hot Encoding on DataFrames.
* **When to use:** Quick EDA and prototyping.
* **Warning:** Avoid in production pipelines because it doesn't remember training categories for test data.

---

## 5. Dummy Variable Trap
* **The Problem:** Keeping all $K$ binary columns causes **multicollinearity** (one column can be predicted by the remaining ones).
* **The Fix:** Retain only **$K-1$ columns** by setting `drop='first'`.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder,OneHotEncoder
```

```python
df=pd.read_csv("/content/tips - tips.csv")
df1=df.copy()
```

```python
print(df1.head())
```

```python
lb=LabelEncoder()
df1['sex']=lb.fit_transform(df1['sex'])
df1['smoker']=lb.fit_transform(df1['smoker'])
df1['day']=lb.fit_transform(df1['day'])
df1['time']=lb.fit_transform(df1['time'])
print(df1.head())
```

```python
df2=df.copy()
oe=OrdinalEncoder(categories=[['Female','Male'],['No','Yes'],['Thur','Fri','Sat','Sun'],['Dinner','Lunch']])
oe_encoded=oe.fit_transform(df2[['sex','smoker','day','time']])
df2[['sex','smoker','day','time']] = oe_encoded
print(df2.head())
```

```python
df3=df.copy()
df3=pd.get_dummies(df3,columns=['sex','smoker','day','time'],drop_first=True,dtype=int)
print("get_dummy", df3.head())
```

```python

df4=df.copy()
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=int)
ohe_col=ohe.fit_transform(df4[['sex','smoker','day','time']])
ohe_encoded=pd.DataFrame(ohe_col)
print(ohe_encoded.head())
```

# Logistic Regression Cheat Sheet

---

## What Is Logistic Regression?
Despite the word "Regression" in its name, Logistic Regression is a **Classification Algorithm** used to predict binary outcomes (e.g., Yes/No, Spam/Not Spam, 0/1).

---

## How It Works (Simple 2-Step Flow)

1. **Calculate Linear Equation:** Calculates a standard line formula:
   $$z = mX + c$$
2. **Squeeze to Probability (Sigmoid):** Passes $z$ into the **Sigmoid Function** to squash any value into a probability between $0$ and $1$:
   $$\sigma(z) = \frac{1}{1 + e^{-z}}$$

* **Decision Rule (Threshold = 0.5):**
  * Probability $\ge 0.5 \rightarrow$ **Class 1 (Yes)**
  * Probability $< 0.5 \rightarrow$ **Class 0 (No)**

---

## Key Metrics to Know

* **Confusion Matrix:** Map of True vs Predicted values (TP, FP, TN, FN).
* **Accuracy:** Overall percentage of correct predictions.
* **Precision:** Out of all positive predictions, how many were actually positive?
* **Recall:** Out of all real positive cases, how many did the model catch?
* **F1-Score:** Balance between Precision and Recall.
* **Log Loss:** The loss function used to train and optimize the model.

```python
import pandas as pd
import numpy as  np
```

```python
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['fever']=df['fever'].fillna(df['fever'].mean())
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['gender']=le.fit_transform(df['gender'])
df['cough']=le.fit_transform(df['cough'])
df['city']=le.fit_transform(df['city'])
df['has_covid']=le.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,lr_pred)
```

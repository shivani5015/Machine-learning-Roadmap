# Navie Bayes :- Naive Bayes is a supervised machine learning algorithm used for classification problems.

##Why is it called Naive Bayes?
Bayes:-

It is based on Bayes Theorem.

P(A|B) = \frac{P(B|A)P(A)}{P(B)}

Naive:-

It assumes that all features are independent of each other.

#Workflow
Training Data

      ↓

Calculate Probabilities

      ↓
Learn Patterns

      ↓

New Data

      
      
Calculate Class Probabilities

      ↓

Choose Highest Probability

      ↓

Prediction

##Types of Naive Bayes:-
1. Gaussian Naive Bayes

Used for continuous numerical data.

from sklearn.naive_bayes import GaussianNB

2. Multinomial Naive Bayes

Used for text and word counts.

3. Bernoulli Naive Bayes

Used for binary features.

#Advantages

✔ Very Fast

✔ Easy to Train

✔ Works well on text data

✔ Requires less training data

✔ Good for spam filtering

#Disadvantages

❌ Assumes features are independent

❌ May perform poorly on complex relationships

❌ Accuracy can be lower than advanced models

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
df.sample(5)
```

```python
df=df.drop(columns=['User ID','Gender'])
```

```python
x=df.drop(columns=['Purchased'])
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_new=sc.fit_transform(x_train)
x_test_new=sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train_new,y_train)
```

```python
y_pred=model.predict(x_test_new)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

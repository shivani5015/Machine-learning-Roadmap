```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import StandardScaler
```

# Regularization Cheat Sheet

---

## What Is Regularization?
A technique used to stop models from **Overfitting**. It penalizes large weight values so the model generalizes better on test data.

$$\text{Total Loss} = \text{MSE Loss} + \text{Penalty}$$

---

## The 3 Types

### 1. Ridge Regression (L2)
* **Penalty:** Adds squared weights ($\alpha \sum w^2$).
* **What it does:** Shrinks weights close to 0, but **never exact 0**.
* **When to use:** Keeps all features; great when features are correlated.

### 2. Lasso Regression (L1)
* **Penalty:** Adds absolute weights ($\alpha \sum |w|$).
* **What it does:** Forces unimportant feature weights to **exact 0**.
* **When to use:** Automatic **Feature Selection** (removes useless features).

### 3. ElasticNet (L1 + L2)
* **Penalty:** Combines both Ridge and Lasso penalties.
* **What it does:** Balances feature selection with correlated feature handling.
* **When to use:** When there are many correlated features and Lasso struggles.

---

## Quick Comparison

| Feature | Ridge (L2) | Lasso (L1) | ElasticNet |
| :--- | :--- | :--- | :--- |
| **Weights = 0?** | No | Yes | Yes |
| **Feature Selection?** | No | Yes | Yes |
| **Main Use** | Prevent Overfitting | Drop Useless Features | Best of Both |

```python
df=pd.read_csv("/content/insurance - insurance.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
```

```python
df.head(2)
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(x_train)
X_test=scaler.transform(x_test)
```

```python
#Ridge
#Ridge adds L2 Penalty
#It reduce coefficient values small as zero but not zero
```

```python
ridge_model=Ridge(alpha=1.0)
ridge_model.fit(X_train,y_train)
```

```python
ridge_pred=ridge_model.predict(x_test)
```

```python
ridge_mse=mean_squared_error(y_test,ridge_pred)
ridge_r2=r2_score(y_test,ridge_pred)
print("MSE",ridge_mse)
print("R2_score",ridge_r2)
print("\ncoeffiecent",ridge_model.coef_)
```

```python
#Lasso adds L1 Penalty
#it can make some coffecient exactly zero to perform feature selection
```

```python
lasso_model=Lasso(alpha=0.1)
lasso_model.fit(x_train,y_train)
```

```python
lasso_pred=lasso_model.predict(x_test)
```

```python
lasso_mse=mean_squared_error(y_test,lasso_pred)
lasso_r2=r2_score(y_test,lasso_pred)
print("MSE",lasso_mse)
print("R2_score",lasso_r2)
print("\ncoeffiecent",lasso_model.coef_)
```

```python
#Elastic Net
 #combines both
 #L1   Regularization (Lasso)
 #L2    Regualrization(Ridge)
```

```python
elastic_model=ElasticNet(alpha=0.1,l1_ratio=0.5)
elastic_model.fit(X_train,y_train)
```

```python
elastic_pred=elastic_model.predict(X_test)
```

```python
elastic_mse=mean_squared_error(y_test,elastic_pred)
elastic_r2=r2_score(y_test,elastic_pred)
print("MSE",elastic_mse)
print("R2_score",elastic_r2)
print("\ncoeffiecent",elastic_model.coef_)
```

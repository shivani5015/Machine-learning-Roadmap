# Quick Cheat-Sheet: Feature Scaling & Train-Test Split

---

## Main Goal
* **Train-Test Split:** Split data into training and testing sets to measure how well the model works on unseen data.
* **Feature Scaling:** Bring all numerical features to the same scale so larger values (e.g., Salary: $50,000$) don't overpower smaller ones (e.g., Age: $25$).

---

## 1. Train-Test Split

* **Common Ratios:** $80/20$ or $70/30$ (Train/Test).
* **`random_state=42`:** Locks the split sequence so you get the exact same train/test data every time you run the code.
* **`stratify=y`:** Preserves the class balance (e.g., $90\%$ normal / $10\%$ fraud) across both train and test sets. Essential for imbalanced datasets.

---

## 2. Feature Scaling Methods

### A. StandardScaler (Standardization)
* **Goal:** Converts data to have a Mean $= 0$ and Standard Deviation $= 1$.
* **When to use:** Distance-based models like KNN, SVM, Linear/Logistic Regression, and Neural Networks.
* **Key Point:** Handles outliers much better than MinMaxScaler.

### B. MinMaxScaler (Normalization)
* **Goal:** Scales all values strictly into a fixed range of $[0, 1]$.
* **When to use:** Algorithms that require bounded ranges (e.g., Image processing, Neural Networks).
* **Key Point:** Highly sensitive to outliers (a single extreme value squashes all normal data into a tiny range).

---

## 3. The Golden Rule (Prevent Data Leakage)
* **On Training Data (`X_train`):** Use `.fit_transform()` (calculates mean/std and transforms the data).
* **On Testing Data (`X_test`):** Use `.transform()` ONLY (applies the same mean/std learned from `X_train`).
* **Why:** The model must never "see" or learn parameters from the test data before evaluation.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler,LabelEncoder
```

```python
df=pd.read_csv("/content/tips - tips.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
lb=LabelEncoder()
df['sex']=lb.fit_transform(df['sex'])
df['smoker']=lb.fit_transform(df['smoker'])
df['day']=lb.fit_transform(df['day'])
df['time']=lb.fit_transform(df['time'])
```

```python
x=df.drop(columns=['tip'])
y=df['tip']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
print(x_train_new.describe().round(2))   #it converts mean=0 and sd=1
```

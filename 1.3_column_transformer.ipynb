# ColumnTransformer Cheat Sheet

---

## Why Use ColumnTransformer?
Real-world datasets contain a mix of feature types (Numerical + Categorical). Applying different preprocessing steps manually and concatenating columns (`np.hstack`) makes code messy and increases the risk of **Data Leakage**.

**`ColumnTransformer`** applies specific transformations to specific column subsets in a single, clean pipeline.

---

## Key Parameters & Concepts

* **Single Pipeline:** Scales numerical features (e.g., `StandardScaler`) and encodes categorical features (e.g., `OneHotEncoder`) simultaneously.
* **`remainder='drop'`:** Drops all columns that were not explicitly transformed (default behavior).
* **`remainder='passthrough'`:** Keeps untransformed columns as they are without modifying them.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/tips - tips.csv")
df.head()
```

```python
#df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['total_bill']),df['total_bill'],test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer()
X_train_num = si.fit_transform(x_train[['tip', 'size']])
X_train_num.shape
```

```python
#ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Female','Male'],['Yes','No'],['Sat','Sun','Thur','Fri']])
x_train_or=oe.fit_transform(x_train[['sex','smoker','day']])
x_test_or=oe.fit_transform(x_test[['sex','smoker','day']])
x_train_or.shape
```

```python
#onehotencoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
x_train_time=ohe.fit_transform(x_train[['time']])
x_test_time=ohe.fit_transform(x_test[['time']])
x_train_time.shape
```

```python
# Merging
x_train_merge = np.concatenate((X_train_num, x_train_or, x_train_time), axis=1)
x_train_merge.shape
```

```python
#by column transformer
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(), ['tip', 'size']),
    ('tnf2', OrdinalEncoder(categories=[['Female', 'Male'],['Yes','No'],['Sat','Sun','Thur','Fri']]), ['sex','smoker','day']),
    ('tnf3', OneHotEncoder(sparse_output=False, drop='first'), ['time'])
], remainder='passthrough')

```

```python
transformer.fit_transform(x_train).shape
```

#  Quick Cheat-Sheet: Decision Tree (Classifier & Regressor)

##  Main Goal
Decision Tree ek non-parametric **Supervised Learning Algorithm** hai jo data ko flowchart-like tree structure mein split karke decisions leta hai.
* **Classifier:** Categorical target variables predict karta hai (e.g., Yes/No).
* **Regressor:** Continuous numerical values predict karta hai (e.g., House Price).

---

##  How Decision Trees Split Data?

### 1. Decision Tree Classifier (Splitting Criteria)

#### A. Gini Impurity (Default in Scikit-Learn)
Purity ki measurement ($0$ matbal 100% pure node). Fast to compute.
$$\text{Gini} = 1 - \sum_{i=1}^{k} p_i^2$$

#### B. Entropy & Information Gain (IG)
Disorder / Uncertainty ki measurement.
$$\text{Entropy}(S) = -\sum_{i=1}^{k} p_i \log_2(p_i)$$
$$\text{Information Gain} = \text{Entropy}(Parent) - \text{Weighted Average Entropy}(Children)$$
* Node us feature par split hota hai jiska **Information Gain Highest** ho.

---

### 2. Decision Tree Regressor (Splitting Criterion)

* **Variance Reduction / MSE:** Numerical target ke liye algorithm variance ya Mean Squared Error (MSE) calculate karta hai.
$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \bar{y})^2$$
* Leaf node ka final prediction: Us node mein aane wale saare data points ka **Average (Mean)**.

---

##  Overfitting & Hyperparameter Tuning (Pruning)

Decision Trees bohot jaldi **Overfit** ho jate hain (Training Accuracy 100% ho jati hai lekin Testing par fail ho jate hain). Isse bachne ke liye Hyperparameters adjust karte hain:

1. **`max_depth`:** Tree ki maximum height/depth limit karna.
2. **`min_samples_split`:** Split karne ke liye minimum kitne samples hone chahiye.
3. **`min_samples_leaf`:** Leaf node par kam se kam kitne samples bachne chahiye.
4. **`max_features`:** Every split par kitne random features consider karne hain.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/insurance - insurance.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
```

```python
df.head(2)
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

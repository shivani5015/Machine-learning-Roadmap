# AdaBoost vs XGBoost Cheat Sheet

---

## What Is Boosting?
A **Sequential Ensemble Technique** where models are built step-by-step. Each new model focuses on correcting the errors made by the previous ones.

---

## 1. AdaBoost (Adaptive Boosting)

* **Core Concept:** Adjusts **sample weights**. It increases weights on misclassified data points so the next model focuses on fixing them.
* **Base Learner:** Uses simple 1-split Decision Trees called **Stumps**.
* **How It Works:**
  1. Assigns equal weights to all data points.
  2. Trains a Decision Stump.
  3. Increases weights for incorrect predictions, decreases weights for correct ones.
  4. Trains the next stump on these updated weights and repeats.

---

## 2. XGBoost (Extreme Gradient Boosting)

* **Core Concept:** Fits models to **Residuals (Errors = Actual - Predicted)** using Gradient Descent.
* **Base Learner:** Uses deeper Decision Trees with built-in Regularization (L1 & L2).
* **How It Works:**
  1. Starts with a base prediction (e.g., mean value).
  2. Calculates residuals (errors).
  3. Trains a new tree to predict those residuals.
  4. Updates predictions: $\text{New Output} = \text{Old Output} + (\text{Learning Rate} \times \text{Residual Prediction})$.

---

## Quick Comparison

| Feature | AdaBoost | XGBoost |
| :--- | :--- | :--- |
| **Error Handling** | Increases weights of missed points | Fits trees to predict Residuals |
| **Base Model** | Decision Stumps (1 split) | Gradient-boosted Trees |
| **Speed & Hardware** | Sequential / Slower | Highly Parallelized / Fast |
| **Regularization** | No built-in L1/L2 | Built-in L1 & L2 (prevents overfitting) |

```python
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x,y=make_classification(n_samples=1000,n_features=10,random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
adaboost =AdaBoostClassifier(n_estimators=100,random_state=42)
adaboost.fit(x_train,y_train)

xgboost= XGBClassifier(n_estimators=100,random_state=42)
xgboost.fit(x_train,y_train)
```

```python
y_pred_adaboost=adaboost.predict(x_test)
y_pred_xgboost=xgboost.predict(x_test)
```

```python
accuracy_adaboost=accuracy_score(y_test,y_pred_adaboost)
accuracy_xgboost=accuracy_score(y_test,y_pred_xgboost)
```

```python
print("adaboost accuracy :",accuracy_adaboost)
print("xgboost accuracy :", accuracy_xgboost)
```

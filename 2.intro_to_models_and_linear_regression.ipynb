# Linear Regression Cheat Sheet

---

## What Is Linear Regression?
A **Supervised Machine Learning Algorithm** used to predict continuous numerical values (e.g., Salary, House Prices, Sales).

---

## Core Formulas

* **Simple Linear Regression (1 Feature):**
  $$y = mX + c$$
  * $y$: Target (Output to predict)
  * $X$: Feature (Input)
  * $m$: Slope / Weight
  * $c$: Intercept / Bias

* **Multiple Linear Regression ($n$ Features):**
  $$y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_n X_n$$

---

## Loss Function & Optimization

* **Loss Function (MSE):** Measures the average squared difference between actual ($y$) and predicted ($\hat{y}$) values.
  $$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$
* **Goal:** Use **OLS** or **Gradient Descent** to adjust weights and minimize MSE to find the best-fit line.

---

## Performance Metrics

* **MAE (Mean Absolute Error):** Average absolute distance from true values. Easiest to interpret.
* **MSE (Mean Squared Error):** Heavily penalizes large errors by squaring them.
* **RMSE (Root Mean Squared Error):** $\sqrt{\text{MSE}}$ — restores errors back to the target variable's original unit.
* **$R^2$ Score:** Measures how much variance in $y$ is explained by the model ($1.0$ = Perfect fit).

---

## Key Assumption

1. **Linearity:** Relationship between features and target must be linear.
2. **No Multicollinearity:** Input features should not be highly correlated with each other (Check using **VIF**).
3. **Homoscedasticity:** Variance of residual errors must stay constant across predictions.
4. **Normality of Residuals:** Model errors should follow a normal distribution.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/insurance - insurance.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
```

```python
df.head()
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
